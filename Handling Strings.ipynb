```python
import pandas as pd
```

```python
df = pd.read_csv('SP.csv')
```

```python
df
```

```python
# How to Handle Special Character
character = ['"', '[' , ']']
for cr in character:
    df['school']=df['school'].str.replace(cr, '')
```

```python
df
```

```python
character = ['as', 'dd' , 'ak']
for cr in character:
    df['score']=df['score'].str.replace(cr, '')
```

```python
df
```

for removing numaric value use '\d+'


```python
df['score'] = df['score'].str.replace('\d+','')
```
